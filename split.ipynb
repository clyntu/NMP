```python
from pathlib import Path
import os
P = Path(os.path.abspath(''))  # Compatible with Jupyter Notebook
P2 = Path('S:\datasets')  # Dataset path
P3 = Path(P / "POP909") # POP909 dataset path
# D = "data/Piano-midi.de"  # Dataset
# D = "data/Nottingham"  # Dataset
# D = "data/JSB Chorales"  # Dataset


def get_file_list_from_dir(datadir):
    all_files = os.listdir(os.path.abspath(datadir))
    data_files = list(filter(lambda file: file.endswith('.mid'), all_files))
    return data_files
```

```python
file_list = []

for x in range(1, 910):
    if x < 10:
        folderName = "00"+str(x)
        file_list.append(get_file_list_from_dir(P3 / folderName))
        continue

    if x < 100:
        folderName = "0"+str(x)
        file_list.append(get_file_list_from_dir(P3 / folderName))
        continue
    
    if x < 910:
        folderName = str(x)
        file_list.append(get_file_list_from_dir(P3 / folderName))
        continue

file_list = [item for sublist in file_list for item in sublist] # flatten the list from 2d to 1d
file_list
```

```python
import random

shuffled_list = file_list
random.seed(230)
random.shuffle(shuffled_list)
```

```python
shuffled_list
```

```python
from math import floor

def get_training_and_testing_sets(file_list):
    split = 0.8
    split_index = floor(len(file_list) * split)
    trainingAndValid = file_list[:split_index]
    testing = file_list[split_index:]
    
    split_training_index = floor(len(trainingAndValid) * split)
    training = trainingAndValid[:split_training_index]
    validation = trainingAndValid[split_training_index:]

    return training, validation, testing
```

```python
train, valid, test = get_training_and_testing_sets(shuffled_list)
```
